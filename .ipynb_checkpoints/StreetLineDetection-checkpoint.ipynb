```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2

#first step to read the video 
#hint : change the path to the target folder 
video = cv2.VideoCapture("solidWhiteRight.mp4")

 #1- read the image
check,image = video.read()
    
while check:
        
    #2 resize the image to specific size
    image = cv2.resize(image, (960,540 )) 
    
    ysize = image.shape[0]
    xsize = image.shape[1]
    region_select = np.copy(image)
    color_select = np.copy(image)

     # Define a kernel size and apply Gaussian smoothing
    kernel_size =5
    blur_image = cv2.GaussianBlur(image,(kernel_size, kernel_size),0)
    
    #convert image to gray scale
    blur_gray = cv2.cvtColor(blur_image,cv2.COLOR_RGB2GRAY)

   #Then read edges of the image
    low_threshold = 150
    high_threshold = 255
    masked_edges = cv2.Canny(blur_gray, low_threshold, high_threshold)


    #crop regions that are not in the boundaries of the street
    # Define the vertices of the mask.
    # WHERE THE LANE LINES ARE IN THE IMAGE
    left_bottom = [120, 540]
    right_bottom = [900, 540]
    top_left = [440, 320]
    top_right = [530, 320]

    #fitting lines of the region
    fit_left = np.polyfit((left_bottom[0], top_left[0]), (left_bottom[1], top_left[1]), 1)
    fit_right = np.polyfit((right_bottom[0], top_right[0]), (right_bottom[1], top_right[1]), 1)
    fit_top = np.polyfit((top_left[0], top_right[0]), (top_left[1], top_right[1]), 1)
    fit_bottom = np.polyfit((left_bottom[0], right_bottom[0]), (left_bottom[1], right_bottom[1]), 1)


    XX, YY = np.meshgrid(np.arange(0, xsize), np.arange(0, ysize))
    region_thresholds = (YY > (XX*fit_left[0] + fit_left[1])) & \
                        (YY > (XX*fit_right[0] + fit_right[1])) & \
                        (YY > (XX*fit_top[0] + fit_top[1])) & \
                         (YY < (XX*fit_bottom[0] + fit_bottom[1]))



    masked_edges[~region_thresholds] = [0]


    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 20     # minimum number of votes (intersections in Hough grid cell)
    min_line_length =5 #minimum number of pixels making up a line
    max_line_gap = 250    # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                                min_line_length, max_line_gap)

    # Iterate over the output "lines" and draw lines on the blank
    for line in lines:
        for x1,y1,x2,y2 in line:
            #P1 = np.array([x1,y1])
            #P2 = np.array([x2,y2])
            #d = linalg.norm(np.cross(P2-P1, P1-P))/linalg.norm(P2-P1)
            #a.append([d,x1,y1,x2,y2])
            cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)


    # Create a "color" binary image to combine with line image
    color_edges = np.dstack((masked_edges, masked_edges, masked_edges)) 

    # Draw the lines on the edge image
    combo = cv2.addWeighted(image, 0.8, line_image, 1, 0) 

    cv2.imshow("combo",combo)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    check,image = video.read()  
        
video.release()        
cv2.destroyAllWindows()        
        



```
